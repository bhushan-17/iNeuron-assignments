```
def myreduce(anyfunc, sequence):


  result = sequence[0]
  
  for item in sequence[1:]:
   result = anyfunc(result, item)

  return result
```

```
def myfilter(anyfunc, sequence):

 
 result = []
 
 for item in sequence:
  if anyfunc(item):
   result.append(item)
 return result
```
