ques.1

Blood glucose levels for obese patients have a mean of 100 with a standard deviation of
15. A researcher thinks that a diet high in raw cornstarch will have a positive effect on
blood glucose levels. A sample of 36 patients who have tried the raw cornstarch diet
have a mean glucose level of 108. Test the hypothesis that the raw cornstarch had an
effect or not.

```
import math
sample_count=36
sample_mean=108
pop_mean=100
pop_std=15
z=(sample_mean-pop_mean)/(pop_std/math.sqrt(sample_count))
print("z score:",z)
```

choosing the value of z from z table 0.9993 hence probability of getting mean less than 108 so 99.93% and for more or eqale is 0.007%,ie less than 0.05% so we will reject null hypotesis

ques.2 

In one state, 52% of the voters are Republicans, and 48% are Democrats. In a second
state, 47% of the voters are Republicans, and 53% are Democrats. Suppose a simple
random sample of 100 voters are surveyed from each state.
What is the probability that the survey will show a greater percentage of Republican
voters in the second state than in the first state?

```
#p1= proportion of republic voters in frist state
#p2=proportion of republic voters in second state
#p1'=proportion of republic voters in sample from the frist state
#p2'=proportion of republic voters in sample from the second state
#N1= the number of voters sampled from the frist state
N1=100
#N2=the number of voters sampled from the second state
N2=100
p1=0.52
#Q1=(1-p1) the proportion on non republic voters in frist state
Q1=0.48
p2=0.47
#Q2=(1-p2) the proportion on non republic voters in second state
Q2=0.53
#the mean of the diference in sample proportion or expected value E[p1'-p2;]

mu=p1-p2
#the standard deviation of the difference (std)
std=math.sqrt(((p1*Q1)/N1)+((p2*Q2)/N2))
print("mu:",mu,"std: ",std)
#here we require to find the probability that p1' is less than p2' ie equvalent to find the probabaility that p1'-p2'<0
x=0
#to find this probabaility, we need to transfer the random  variable (p1'-p2')into a z-score
#Z=Z_score(p1',p2')
#that transformation appears below
Z=(x-mu)/std
print("Z_score(p1',p2':", Z)
#from Z table we find that the probability of a Z-score being -0.7082 or less is 0.24
print('thr probability that the survey will show a greater percentage of re[public voters in second state that in frist state is 0.24')
```

ques.3

You take the SAT and score 1100. The mean score for the SAT is 1026 and the standard
deviation is 209. How well did you score on the test compared to the average test taker?

```
#the z score tells you how many std from the mean your score is my score=x
x=1100
#population means =mu
mu=1026
#population std =sd
sd=209
z=(x-mu)/sd
print("Z score:",z)
print('the above calculation shows that my score is 0.35 std above the mean ')
```
